```python
#pandas is a library where it contains various functions relating to database 
import pandas as pd
```

```python
#dataset is called using the variable (read_csv)by the library function acting as class and specific dataset considered is called
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#now check the data in dataset
dataset
```

```python
#considering independent as i/p variable and consider yr of exp as input and display it
independent=dataset[["YearsExperience"]]
```

```python
#check the data in the variable
independent
```

```python
#considering dependent as o/p variable and consider salary as output and display it
dependent=dataset[["Salary"]]
```

```python
#check the data in the variable
dependent
```

```python
#separate or split train and test set by using function call(train_test_split)
from sklearn.model_selection import train_test_split
#function loaded and consider as parameterized function and use the i/p and o/p of the model
#size of the test(training size to be higher and use random state)
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
#it displays a list with 4 (i/p and o/p) so need to assign var to use in another function, hence x_train,x_test & y_train,y_test
```

```python
#using algorithm in python,create model using linear regression-Procedure
from sklearn.linear_model import LinearRegression
#var regressor cosidered and linear regression considered as class- Procedure
regressor=LinearRegression()
#call fit method/fun using access operator to make learn using material (X_train -i/p & y_train- o/p)
regressor.fit(X_train,y_train)
```

```python
#model created using linear reg(weight & intercept) and need to know the calc value to check it
#coef method weight is calculated
weight=regressor.coef_
#the slope is hook for next prediction
weight
```

```python
#bias method interept is calculated
bias=regressor.intercept_
bias
```

```python
#evaluation metric, prediction is application of the function,evaluation to be done only for test
#cross check for brain prediction with the test
y_pred=regressor.predict(X_test)
```

```python
#calculate R2
from sklearn.metrics import r2_score
#create var name R2score(actual- y_test,predicted -y_pred)
r_score=r2_score(y_test,y_pred)
```

```python
#r_score if 0 ,poor mdel and if 1 its good model
r_score
```

```python
#saving the model using function name pickle
import pickle
#creating dummy filename for finalized model with algm name use .sav for saving
filename="finalized_model_linear.sav"
```

```python
#assign model under function,wb-write binary
pickle.dump(regressor,open(filename,'wb' ))
```

```python
#saved model to load for checking,create var name loaded_model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#real time test by sending value
result=loaded_model.predict([[13]])
```
